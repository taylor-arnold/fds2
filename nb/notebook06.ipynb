**Answer**:
Hurricane Center recycles names on a six-year rotation, so a name is reused

identify a storm. The name and the year together do.

**Answer**:
the category is not a second measurement of the storm. It is calculated from the
wind speed by looking up which band the speed falls into: 65 to 80 knots is a
category 1, 140 knots and above is a category 5, and anything under 35 knots is
recorded as a -1, meaning the system was not yet a tropical storm. A plot of two
columns usually tells you how two things relate to each other in the world. Here
it tells you the definition of a word. It is worth knowing which of your columns
were measured and which were derived from the others, because only the first
kind can surprise you.
